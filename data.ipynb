```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import ipywidgets as widgets
from ipywidgets import interact

def graphique_interactif(N):
    # Normal distribution parameters
    mu, sigma = 0, 1

    # Data generation
    data = np.random.normal(mu, sigma, int(N))

    # Plot creation
    plt.figure(figsize=(10, 6))

    # Histogram
    plt.hist(data, bins=50, density=True, alpha=0.5, color='royalblue', label=f'Samples (N={N})')

    # Theoretical curve
    x = np.linspace(-4, 4, 1000)
    plt.plot(x, norm.pdf(x, mu, sigma), 'r-', lw=2, label='Theory (Normal Distribution)')

    # Formatting
    plt.title(f"Convergence to Normal Distribution for $N = {N}$", fontsize=14)
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.ylim(0, 0.6)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

# Creating the interactive slider for Google Colab
interact(graphique_interactif,
         N=widgets.IntSlider(min=500, max=25000, step=10, value=1000, description='Sample Size N:'));
```
